```python

# a) Binary Search
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = low + (high - low) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

ordered_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]
target = 5
result = binary_search(ordered_list, target)
if result != -1:
    print(f"Item found at index: {result}")
else:
    print("Item not found.")
```

```python
# b) Binary Search using Recursion
def recursive_binary_search(arr, target, low, high):
    if low > high:
        return -1
    mid = low + (high - low) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return recursive_binary_search(arr, target, mid + 1, high)
    else:
        return recursive_binary_search(arr, target, low, mid - 1)

ordered_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]
item_to_find = 5
result = recursive_binary_search(ordered_list, item_to_find, 0, len(ordered_list) - 1)
if result != -1:
    print(f"Item found at index: {result}")
else:
    print("Item not found.")
```
