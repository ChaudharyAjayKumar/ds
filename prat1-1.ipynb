```python
def linear_search(lst, target):
    for i in range(len(lst)):
        if lst[i] == target:
            return i  # Return the index if found
    return -1  # Return -1 if not found

my_list = [4, 2, 7, 1, 9]
target_item = 7
result = linear_search(my_list, target_item)

if result != -1:
    print(f"Item found at index: {result}")
else:
    print("Item not found.")
```

```python
# b) Searching for a substring
def search_substring_in_file(file_path, substring):
    with open(file_path, 'r') as file:
        for line_number, line in enumerate(file, start=1):
            if substring in line:
                print(f"Found '{substring}' in line {line_number}: {line.strip()}")

file_path = 'example.txt'
substring_to_search = 'target'
search_substring_in_file('/content/file.txt', "hello")
```

```python
# c) Finding an element in a NumPy array
import numpy as np

def check_element_in_array(arr, target):
    return target in arr

array = np.array([1, 2, 3, 4, 5])
target_element = 3

if check_element_in_array(array, target_element):
    print(f"Element {target_element} is in the array.")
else:
    print(f"Element {target_element} is not in the array.")
```

```python
# d) Searching for a value in Pandas DataFrame
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40]
}
df = pd.DataFrame(data)

search_value = 'Bob'
result = df[df['Name'] == search_value]

if not result.empty:
    print("Found:")
    print(result)
else:
    print(f"{search_value} not found in the DataFrame.")
```
