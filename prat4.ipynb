```python
# a) Set Operations
import csv

def load_books(filename):
    books_set = set()
    with open('/content/Book1.csv', 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            book = (row["Author"], row["Title"])
            books_set.add(book)
    return books_set

def display(books_set, description):
    print(description)
    for book in books_set:
        print(f"Author: {book[0]} Title: {book[1]}")

books = load_books("Book1.csv")
display(books, "Here are the books")

books.add(("Author4", "Title4"))
display(books, "Books after adding")
books.discard(("Author4", "Title4"))
display(books, "Books after discarding")

books1 = load_books("Book2.csv")
display(books1, "Here are books2")

# Union
books3 = books.union(books1)
display(books3, "Union")

# Intersection
books3 = books.intersection(books1)
display(books3, "Intersection")

# Difference
books3 = books - books1
display(books3, "Difference")

# Symmetric Difference
books3 = books1.symmetric_difference(books)
display(books3, "Symmetric Difference")
```
